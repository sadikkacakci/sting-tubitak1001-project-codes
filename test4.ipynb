```python
from dummy_data_generator import DummyDataGenerator
import matplotlib.pyplot as plt
```

```python
treatment_process = 90
number_of_patients = 12
generator = DummyDataGenerator(treatment_process,number_of_patients)
```

```python
df = generator.get_dummy_data()
```

```python
temp = df.loc[df["Patient_ID"] == 1000]
```

```python
# 6MP gnlk doz ve MTX haftalk dozunun grafiini iz
plt.figure(figsize=(10, 5))
plt.plot(temp['Date'], temp['6MP_Daily_Dose_mg'], label='6MP Daily Dose (mg)', marker='o',linestyle='None')
plt.plot(temp['Date'], temp['MTX_Weekly_Dose_mg'], label='MTX Weekly Dose (mg)', marker='s',linestyle='None')
plt.title('6MP Daily Dose vs MTX Weekly Dose')
plt.xlabel('Date')
plt.ylabel('Dose (mg)')
plt.ylim(0, 90)  # Y ekseninin limitlerini 0 ile 90 arasnda ayarla
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# 6MP gnlk doz ve MTX haftalk dozunun grafiini iz
plt.figure(figsize=(10, 5))
plt.plot(temp['Date'], temp['ANC_G_L'], label='ANC_G_L',marker='o',linestyle='',color="gray")
plt.title('ANC_G_L')
plt.xlabel('Date')
plt.ylabel('Dose (mg)')
plt.ylim(0, 8)  # Y ekseninin limitlerini 0 ile 90 arasnda ayarla
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
